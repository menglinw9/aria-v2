```python
# 3. 讀取資料 (請修改為實際檔案路徑)
file_path_dem = '/content/drive/MyDrive/GIS_data/dem_20m_hualien.tif'
file_path_boundary = '/content/drive/MyDrive/GIS_data/hualien_buffer.geojson'

# Check if files exist before attempting to open them
if not os.path.exists(file_path_dem):
    raise FileNotFoundError(f"錯誤：DEM 檔案未找到於 {file_path_dem}。請確認檔案已上傳至您的 Google 雲端硬碟或更新路徑。")
if not os.path.exists(file_path_boundary):
    raise FileNotFoundError(f"錯誤：邊界檔案未找到於 {file_path_boundary}。請確認檔案已上傳至您的 Google 雲端硬碟或更新路徑。")

dem = rioxarray.open_rasterio(file_path_dem)
boundary = gpd.read_file(file_path_boundary)

# 4. 統一座標系統 (CRS) 以防錯誤
boundary = boundary.to_crs(dem.rio.crs)

# 5. 執行裁切 (drop=True 代表去除邊界外的無效數值空間)
clipped_dem = dem.rio.clip(boundary.geometry, boundary.crs, drop=True)

# 6. 存回 Google Drive
#clipped_dem.rio.to_raster('/content/drive/MyDrive/clipped_dem.tif')
```
